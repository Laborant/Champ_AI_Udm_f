```python
from operator import itemgetter
from os.path import join
from tqdm import tqdm

import os
import copy
import json

path_train = "" # path to train images
path_ann = "" # path to full anootation

def get_json(path):
    
    json_file = open(path)
    coco_json = json.load(json_file)
    json_file.close()
    
    return coco_json
```

```python
data = get_json(path_ann)
res_data = copy.deepcopy(data)
```

```python
list_imgs = sorted(os.listdir(path_train))
list_imgs
```

```python
def get_list_ann(_json, find_image_id):
    list_ann = []
    for idx, val in enumerate(_json["annotations"]):
        if val["image_id"] == find_image_id:
            list_ann.append(_json["annotations"][idx])
    return list_ann
```

```python

data["images"] = sorted(data["images"], key=itemgetter('file_name')) 
```

```python
data["images"]
```

```python
data["annotations"] = sorted(data["annotations"], key=itemgetter('id')) 
```

```python
data["annotations"]
```

```python
ann_list = []
img_list = []
new_image_id = 1
new_id = 1

for i in tqdm(range(0, len(data["images"]))):
    if data["images"][i]['file_name'] in list_imgs:

        _list_ann = get_list_ann(data, data["images"][i]['id'])

        data['images'][i]['id'] = new_image_id
        img_list.append(data['images'][i])
        
        for idx, _ in enumerate(_list_ann):
            _list_ann[idx]['image_id'] = new_image_id
            _list_ann[idx]['id'] = new_id
            new_id += 1
        
        if _list_ann:
            ann_list += _list_ann
        
        new_image_id += 1
        
        if len(img_list) == 2100:
            break
```

```python
res_data["images"] = img_list
res_data["annotations"] = ann_list
```

```python
get_list_ann(res_data, 2100)
```

```python
with open("annotation-train.json", "w") as outfile:
    json.dump(res_data, outfile)
```
